```python
# import required libraries
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# read the html file as soup object
with open('practice-exercise-1.html', 'r') as f:
    soup = BeautifulSoup(f)
```

```python
# print the title of the webpage
print(soup.title)
```

```python
# print the head of the webpage using perttify()
head = soup.find('thead')
print(head.prettify())
```

```python
# print the body of the webpage using perttify()
body = soup.find('tbody')
print(body.prettify())
```

```python
# get the column names from the head and store it as a string
columns = [x.text for x in head.select('th')]
columns
```

```python
# iterate through each row and create a list of values and then append those lists into a final list 
# hint: each row is under a <tr> tag, and each values is under a <td> tag 
# use the .find method and .text method to extract your data
data = []
rows = body.select('tr')

for row in rows:
    model = row.find('td', class_='model').text
    chip = row.find('td', class_='chip').text
    release_date = row.find('td', class_='release-date').text
    bus = row.find('td', class_='bus').text
    memory = row.find('td', class_='memory').text
    gpu_clock = row.find('td', class_='gpu-clock').text
    memory_clock = row.find('td', class_='memory-clock').text
    cores = row.find('td', class_='cores').text
    data.append([model, chip, release_date, bus, memory, gpu_clock, memory_clock, cores])
```

```python
# verify if you got everything using the length function on data
len(data)
```

```python
# convert the list into a dataframe by specifiy the list and the column names
df = pd.DataFrame(data=data, columns=columns)
```
