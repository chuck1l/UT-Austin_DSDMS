```python
# import required libraries
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# read the html file as soup object
with open('practice-exercise-2.html', 'r') as f:
    soup = BeautifulSoup(f)
```

```python
# print the title of the webpage
print(soup.title)
```

```python
# print the head of the webpage using perttify()
head = soup.find('head')
print(head.prettify())
```

```python
# print the body of the webpage using perttify()
body = soup.find('body')
print(body.prettify())
```

```python
# find all the laptops using the finaAll method.
# hint: look for class "item-cell"
laptops = soup.findAll('div', attrs={'class':'item-cell'})
len(laptops)
```

```python
# print the html text of a single laptop
print(laptops[0].prettify())
```

```python
# iterate through each card and create a list of values and then append those lists into a final list 
# hint: each card is under a <div> tag, and all the features are under a <li> tag 
# use the .find method and .text method to extract your data
data = []
laptops = soup.findAll('div', attrs={'class':'item-cell'})

for laptop in laptops:
    brandname = laptop.find('div', class_='brand-name').text
    title = laptop.find('div', class_='item-title').text
    price = laptop.find('strong').text
    features = [x.text for x in laptop.select('li')]
    graphic, resolution, weight = features
    data.append([brandname, title, price, graphic, resolution, weight])   
```

```python
# verify if you got everything using the length function on data
len(data)
```

```python
# convert the list into a dataframe by specifiy the list and the column names
columns = ['brand', 'title', 'price', 'graphic', 'resolution', 'weight']
df = pd.DataFrame(data=data, columns=columns)
```
